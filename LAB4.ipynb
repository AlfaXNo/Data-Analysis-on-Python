Брошены три монеты. Найти $P(A)$, $P(B)$, $P(C)$, где:

а) $A = \{\text{первая монета выпала гербом}\}$

б) $B = \{\text{ровно два герба}\}$

в) $C = \{\text{не больше двух гербов}\}$

\section*{Брошены три монеты}

\textbf{Пространство элементарных исходов:} 
$\Omega = \{(Г,Г,Г), (Г,Г,Р), (Г,Р,Г), (Г,Р,Р), (Р,Г,Г), (Р,Г,Р), (Р,Р,Г), (Р,Р,Р)\}$ \\
Всего исходов: $n = 8$

а) Событие $A = \{\text{первая монета выпала гербом}\}$

Благоприятные исходы: 
$$(Г,Г,Г), (Г,Г,Р), (Г,Р,Г), (Г,Р,Р)$$
Количество благоприятных исходов: $m_A = 4$

Вероятность:
$$P(A) = \frac{m_A}{n} = \frac{4}{8} = \frac{1}{2}$$

б) Событие $B = \{\text{выпало ровно два герба}\}$

Благоприятные исходы:
$$(Г,Г,Р), (Г,Р,Г), (Р,Г,Г)$$
Количество благоприятных исходов: $m_B = 3$

Вероятность:
$$P(B) = \frac{m_B}{n} = \frac{3}{8}$$

в) Событие $C = \{\text{выпало не больше двух гербов}\}$

Благоприятные исходы (все, кроме трёх гербов):
$$(Г,Г,Р), (Г,Р,Г), (Г,Р,Р), (Р,Г,Г), (Р,Г,Р), (Р,Р,Г), (Р,Р,Р)$$
Количество благоприятных исходов: $m_C = 7$

Вероятность:
$$P(C) = \frac{m_C}{n} = \frac{7}{8}$$

Ответ:

$P(A) = \frac{1}{2} = 0.5$,

$P(B) = \frac{3}{8} = 0.375$, 

$P(C) = \frac{7}{8} = 0.875$

```python
heads_first = 0
two_heads = 0
less_three_heads = 0
n = 100000
result = [0,1] # 1 - орёл, 0 - решка
for i in range(n):
    m1 = random.choice(result)
    m2 = random.choice(result)
    m3 = random.choice(result)
    summ = m1+m2+m3
    if m1 == 1:
        heads_first += 1
    if summ == 2:
        two_heads += 1
    if summ <= 2:
        less_three_heads += 1

print(f"P(A) = {heads_first/n}")
print(f"P(B) = {two_heads/n}")
print(f"P(C) = {less_three_heads/n}")
```

Бросаются две игральные кости. Пусть событие $A$ состоит в том,
что выпавшая сумма очков нечётна, а событие $B$ — в том, что хотя бы
на одной из костей выпала единица. Описать события:

а) $A \cap B$

б) $\overline{A \cap B}$

в) $A \cup B$

г) $A \cap \overline{B}$

Решение

Пространство элементарных исходов:

При бросании двух игральных костей всего возможных исходов:
$$n = 6 \times 6 = 36$$

Исходные события:

$A = \{\text{сумма очков нечётна}\}$ 

$B = \{\text{хотя бы на одной кости выпала единица}\}$

а) $A \cap B$

Описание: Сумма нечётна И хотя бы одна единица.

Благоприятные исходы:
$$(1,2), (1,4), (1,6), (2,1), (4,1), (6,1)$$
$$m_{A \cap B} = 6$$

Вероятность:
$$P(A \cap B) = \frac{6}{36} = \frac{1}{6}$$

б) $\overline{A \cap B}$

Описание: Сумма чётна ИЛИ нет единицы.

Отрицание пункта а:

Вычисляем по  формуле:

$P(\overline{A})= 1 - P(A)$

$P(\overline{A \cap B}) = 1- \frac{1}{6} = \frac{5}{6}$


в) $A \cup B$

Описание: Сумма нечётна ИЛИ хотя бы одна единица.

Вычисляем по формуле:
$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

Вероятности:
$P(A) = \frac{18}{36} = \frac{1}{2}$ (нечётная сумма: 18 исходов)

$P(B) = \frac{11}{36}$ (хотя бы одна единица: 11 исходов)

$P(A \cap B) = \frac{6}{36} = \frac{1}{6}$ (из пункта а)


Вероятность:
$$P(A \cup B) = \frac{18}{36} + \frac{11}{36} - \frac{6}{36} = \frac{23}{36}$$

г) $A \cap \overline{B}$

Описание: Сумма нечётна И ни на одной кости нет единицы.

$$P(A \cap \overline{B}) = P(A) - P(A \cap B) = \frac{18}{36} - \frac{6}{36} = \frac{12}{36} = \frac{1}{3}$$

Ответ:

а) $P(A \cap B) = \frac{1}{6} = 0.167$

б) $P(\overline{A} \cap B) = \frac{5}{6} = 0.833$

в) $P(A \cup B) = \frac{23}{36} = 0639$

г) $P(A \cap \overline{B}) = \frac{1}{3} = 0.333$


```python
a_and_b = 0
not_a_or_not_b = 0
a_or_b = 0
a_and_not_b = 0
n = 100000
for i in range(n):
    a = random.randint(1,6)
    b = random.randint(1,6)
    summ = a + b
    if summ % 2 != 0 and (a == 1 or b == 1):
        a_and_b+=1
    if summ % 2 == 0 or (a != 1 and b != 1):
        not_a_or_not_b += 1
    if summ % 2 != 0 or a == 1 or b == 1:
        a_or_b += 1
    if summ % 2 != 0 and (a != 1 and b != 1):
        a_and_not_b += 1

print(f"а) {a_and_b/n}")
print(f"б) {not_a_or_not_b/n}")
print(f"в) {a_or_b/n}")
print(f"г) {a_and_not_b/n}")
```

#### Задача Бюффона о бросании иглы 

Пусть $x$ - расстояние от центра иглы до ближайшей прямой ($0 \leq x \leq a$), а $\phi$ - угол между иглой и прямыми ($0 \leq \phi \leq \pi/2$).

Игла пересечёт прямую, если 
$$x \leq r \sin \phi$$

где $r$ - длина иглы, $a$ - расстояние между параллельными прямыми.

*Вероятность пересечения*

Вероятность того, что игла пересечёт прямую, равна отношению благоприятной площади к общей площади в пространстве параметров:

$$P = \frac{\int_0^{\pi/2} \int_0^{r \sin \phi} dx\, d\phi}{\int_0^{\pi/2} \int_0^{a} dx\, d\phi}
= \frac{\int_0^{\pi/2} r \sin \phi\, d\phi}{\int_0^{\pi/2} a\, d\phi}
= \frac{r[- \cos \phi]_0^{\pi/2}}{a \cdot \frac{\pi}{2}}
= \frac{2r}{\pi a}$$

```python
n = 100000
count = 0
a = 100
r = 50
for i in range(n):
    x = random.uniform(0, a)
    phi = random.uniform(0, math.pi / 2)
    if x <= r * math.sin(phi):
        count += 1

print(f"Вероятность по методу Монте-Карло: {count/n}")
print(f"Ответ: {(2 * r) / (math.pi * a):.5f}")
```
